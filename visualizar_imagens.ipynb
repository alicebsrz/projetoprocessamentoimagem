```python
# ===================================================================
# Bloco de Cdigo para o VS Code com a extenso Jupyter
# Nome do arquivo: visualizar_imagens.ipynb
# ===================================================================

# Passo 1: Importar as bibliotecas necessrias
import matplotlib.pyplot as plt
import glob
import skimage.io as io
import os

# Passo 2: Definir o caminho para a pasta de imagens
# Como abrimos a pasta do projeto, o caminho relativo 'Imagens/*.jpg' funciona perfeitamente.
path_imagens = 'Imagens/*.jpg'

# Passo 3: Listar todas as imagens encontradas no caminho
lista_de_imagens = sorted(glob.glob(path_imagens))

print(f"Total de imagens encontradas: {len(lista_de_imagens)}")

# Passo 4: Configurar a grade de visualizao (plot)
# Grade 4x4 para as 16 imagens
fig, axes = plt.subplots(4, 4, figsize=(10, 10))
fig.suptitle('Base de Imagens da Turma', fontsize=16)

# Passo 5: Percorrer a lista de imagens e exibi-las na grade
for ax, arquivo_imagem in zip(axes.ravel(), lista_de_imagens):
    imagem = io.imread(arquivo_imagem)
    ax.imshow(imagem)

    # Extrai o nome da classe do nome do arquivo para usar como ttulo
    nome_arquivo = os.path.basename(arquivo_imagem)
    titulo = nome_arquivo.split('_')[0]
    ax.set_title(titulo)

    ax.axis('off') # Remove os eixos para uma visualizao limpa

# Passo 6: Ajustar o layout e mostrar o grfico
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
